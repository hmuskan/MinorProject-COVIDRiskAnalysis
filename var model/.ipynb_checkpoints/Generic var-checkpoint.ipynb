```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import seaborn as sns
%matplotlib inline
import plotly.express as px
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, acf, pacf,arma_order_select_ic
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tools.eval_measures import rmse, aic
import warnings
warnings.simplefilter('ignore')
import joblib
import os

```

```python
def var(train,st,dis):
    model=VAR(train)
    fitted=model.fit(maxlags=12,ic='aic')
    ##Exporting model
    modelname="varmodel_for_state_"+st+"_district_"+dis
    print(modelname)
    joblib.dump(fitted,modelname)

```

```python
list = ['Date', 'State', 'District', 'Confirmed','Recovered','Deceased']
df= pd.read_csv('output-Copy1.csv', usecols=list)
df
```

```python

df['Date']=pd.to_datetime(df['Date'])
df['Confirmed log'] = np.log(df['Confirmed'])
df['Recovered log'] = np.log(df['Recovered'])
df['Deceased log'] = np.log(df['Deceased']) 
df=df[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
dftemp=df.copy()
df=df.drop(['Confirmed','Recovered','Deceased'],axis=1)
df
```

```python
mpe=pd.read_csv('Mapescoreinfo.csv')
mpe.rename(columns = {'test':'TEST'}, inplace = True)
mpe
```

```python

state_dist=set(zip(df['State'],df['District']))
for [s,d] in state_dist:
    modelname="varmodel_for_state_"+s+"_district_"+d
    if modelname in os.listdir():
        continue
    else:     
        try:
            
            dftemp=df[df['District']==d]
            train=dftemp[:120]
            test=dftemp[120:]
            train = train.set_index(['Date'])
            test = test.set_index(['Date'])
            traintemp=train.loc[:,['Confirmed log','Recovered log','Deceased log']]
            testtemp=test.loc[:,['Confirmed log','Recovered log','Deceased log']]
            var(traintemp,s,d)
            m=joblib.load(modelname)
            # Get the lag order
            lag_order = m.k_ar

            # Input data for forecasting
            forecast_input = testtemp.values[-lag_order:]
            forecast_input
            pred=m.forecast(forecast_input ,len(test.dropna()))
            test['predictedconfirmedlog']=pred[:,0]
            test['recoveredconfirmedlog']=pred[:,1]
            vl=mape(test['Confirmed log'],test['predictedconfirmedlog'])
            mpe.at[mpe['District']==d,'MapeScore var']=vl
        except:
            continue

```

```python
def mape(y2, y_pred): 
    y2, y_pred = np.array(y2), np.array(y_pred)
    return np.mean(np.abs((y2 - y_pred) / y2)) * 100
print('\nMean absolute percentage error: %f'%mape(test['Confirmed log'],test['predictedconfirmedlog']))
```
